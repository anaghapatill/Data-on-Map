## CITYSCAPES

#### Loading the train split of **Cityscapes(leftImg8bit)** dataset. The dataset can be downloaded from https://www.cityscapes-dataset.com/login/ after registering/logging in.

```python
cityscapes_dataset = foz.load_zoo_dataset(
    "cityscapes",
    split="train",
)
```

##### Train path variable is the path to the directory where the JSONs are stored.

```python
TRAIN_PATH = "/home/wln2kor/Downloads/vehicle_trainvaltest/vehicle/train/"
for sample in cityscapes_dataset:
    filename = sample["filepath"].split("/")[-1][:-16]
    city = filename.split("_")[0]
    json_file = TRAIN_PATH + city + "/" + filename + "_vehicle.json"
    f = open(json_file)
    json_data = json.load(f)
    for k,v in json_data.items():
        sample[k] = v
        sample.save()
    sample["GPS_Location"] = fo.core.labels.GeoLocation(point=[sample["gpsLongitude"],sample["gpsLatitude"]])
    sample.save()
cityscapes_dataset.save()
```

```python
model = foz.load_zoo_model("faster-rcnn-resnet50-fpn-coco-torch")
cityscapes_dataset.apply_model(model, label_field="predictions")
```

```python
mean_confidence = [0 if(len(x)==0) else sum(x)/len(x) for x in cityscapes_dataset.values("predictions.detections.confidence")]
num_objects = [len(x) for x in cityscapes_dataset.values("predictions.detections.confidence")]
```

```python
session = fo.launch_app(cityscapes_dataset,port = 5161)
```

```python
plot = fo.location_scatterplot(
    samples=cityscapes_dataset,
    labels=mean_confidence,
    sizes=num_objects,      # scale point sizes by number of objects
    labels_title="Mean Confidence",
    sizes_title="Number of objects",
)
plot.show(height=720)
session.plots.attach(plot)
```
